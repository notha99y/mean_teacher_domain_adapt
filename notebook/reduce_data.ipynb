```python
import os
os.chdir(os.path.join(os.getcwd(), '..'))
```

```python
from shutil import copyfile
import random
```

```python
reduce_data_path = os.path.join(os.getcwd(), 'reduced_data')
full_data_path = os.path.join(os.getcwd(), 'data')
```

```python
NUM_OF_IMAGES_PER_CLASS = 1000
TYPE = 'real'
```

```python
reduce_class = ['aeroplane', 'car', 'skateboard']
```

```python
walker = os.walk(os.path.join(full_data_path, TYPE))
next(walker)
for r,d,f in walker:
#     print(r.split('/')[-1])
    if r.split('/')[-1] in reduce_class:
        save_path = os.path.join(reduce_data_path,TYPE, r.split('/')[-1])
        os.makedirs(save_path)
        while len(os.listdir(save_path)) < NUM_OF_IMAGES_PER_CLASS:
            randomly_choosen = random.sample(f, 1)[0]
            print(randomly_choosen)
            copyfile(os.path.join(r, randomly_choosen), os.path.join(save_path,randomly_choosen))
    
```
